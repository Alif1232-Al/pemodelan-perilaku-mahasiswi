# 🌐 Analisis Perilaku Seksual Berisiko Mahasiswi Era Digital: Dekonstruksi Riset Klasik Berbasis Teori Siklus Trauma & FoMO

### 📌 Latar Belakang & Pembaruan Konteks (Novelty)
Proyek data science ini dirancang sebagai bentuk pembaruan (*context shift*) dan kritik akademis terhadap studi perilaku seksual mahasiswa tradisional (seperti riset klasik *Khasanah, N., 2016* di Universitas Negeri Malang).

Jika penelitian di masa lalu hanya melihat perilaku seksual berpacaran dari kacamata hitam-putih seperti tingkat religiusitas dan kontrol diri internal, pemodelan data kontemporer (Era 2026) ini mengintegrasikan variabel eksternal yang jauh lebih destruktif: **Digital Normalization (Normalisasi Digital melalui FYP Media Sosial)** dan **FoMO-Driven Consumptive Lifestyle (Tekanan Gaya Hidup Konsumtif)**.

---

### 🧠 Kerangka Hipotesis yang Diuji (The Hypothesis Cycle)
Analisis ini berfokus pada sub-populasi spesifik: **Mahasiswi Baru (Perantau) yang tinggal di rumah kost dan telah aktif dalam lingkaran pergaulan bebas**. Data di dalam proyek ini disimulasikan untuk menguji 3 fase hipotesis psikososial berikut:

1. **Fase Desensitisasi Digital (Algoritma Medsos):** Paparan konstan terhadap konten *glorifikasi relationship* di media sosial mengikis batasan tabu. Seks pranikah yang dulunya dianggap menyimpang bergeser menjadi konsumsi umum dan diwajarkan sebagai bagian dari tren gaya hidup.
2. **Fase Eksploitasi & Tuntutan FoMO (Motivasi Transaksional):** Hilangnya kontrol sosial primer (orang tua) di lingkungan kos berbenturan dengan kecemasan sosial (*Fear of Missing Out*). Mahasiswi dengan keterbatasan finansial rentan terjebak dalam hubungan yang bergeser menjadi semi-transaksional demi mendanai gengsi gaya hidup digitalnya (skincare, fashion, tempat nongkrong viral).
3. **Fase Siklus Trauma Bercabang (*Repetition Compulsion*):** Pengalaman pertama (*first sex*) yang sering kali didasari manipulasi pasangan memicu depresi klinis dan kehancuran harga diri (*low self-esteem*). Alih-alih berhenti, ketiadaan *support system* memaksa korban masuk ke dalam mekanisme koping yang salah (*self-destructive*). Aktivitas seksual diulangi terus-menerus karena kondisi mental yang sudah mati rasa psikologis ("terlanjur basah"), hingga polanya bercabang menjadi perilaku berisiko yang akut.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Set seed biar data konsisten pas dijalankan
np.random.seed(77)
n_samples = 60 # Gua tambahin sampelnya jadi 60 biar variasi datanya makin cakep

# 2. Definisi variabel pilihan
status_tinggal_pilihan = ['Kost', 'Rumah Bersama Ortu']
aturan_jam_pilihan = ['Tidak Ada Aturan', 'Ada Aturan/Ketat']
orientasi_pilihan = ['Jangka Pendek (Impulsif)', 'Jangka Panjang (Prudent)']
kencan_instan_pilihan = ['Mudah/Cepat Diajak', 'Sulit/Selektif']

data_rows = []
for i in range(n_samples):
    id_resp = f'Remaja_{i+1}'
    usia = np.random.randint(16, 22)
    gender = np.random.choice(['Laki-laki', 'Perempuan'], p=[0.35, 0.65]) # Fokus utama tetap populasi perempuan

    if gender == 'Perempuan':
        status_tinggal = np.random.choice(status_tinggal_pilihan, p=[0.5, 0.5])
        aturan_jam = np.random.choice(aturan_jam_pilihan, p=[0.5, 0.5])

        # Logika Inti: JIKA KOST + TIDAK ADA JAM MALAM -> Cenderung berorientasi Jangka Pendek (Impulsif)
        if status_tinggal == 'Kost' and aturan_jam == 'Tidak Ada Aturan':
            orientasi_masa_depan = np.random.choice(orientasi_pilihan, p=[0.85, 0.15])
        else:
            orientasi_masa_depan = np.random.choice(orientasi_pilihan, p=[0.4, 0.6])

        # Logika Tambahan: JIKA Impulsif/Jangka Pendek -> Sangat mudah diajak kencan instan oleh laki-laki muda
        if orientasi_masa_depan == 'Jangka Pendek (Impulsif)':
            kencan_instan = np.random.choice(kencan_instan_pilihan, p=[0.90, 0.10]) # 90% mudah diajak kencan
            perilaku_bebas = np.random.choice(['Ya', 'Tidak'], p=[0.85, 0.15])
        else:
            kencan_instan = np.random.choice(kencan_instan_pilihan, p=[0.30, 0.70])
            perilaku_bebas = np.random.choice(['Ya', 'Tidak'], p=[0.20, 0.80])

    else:
        # Logika standar pengisi data untuk populasi Laki-laki
        status_tinggal = np.random.choice(status_tinggal_pilihan, p=[0.4, 0.6])
        aturan_jam = np.random.choice(aturan_jam_pilihan, p=[0.4, 0.6])
        orientasi_masa_depan = np.random.choice(orientasi_pilihan, p=[0.5, 0.5])
        kencan_instan = np.random.choice(kencan_instan_pilihan, p=[0.6, 0.4])
        perilaku_bebas = np.random.choice(['Ya', 'Tidak'], p=[0.45, 0.55])

    data_rows.append([id_resp, usia, gender, status_tinggal, aturan_jam, orientasi_masa_depan, kencan_instan, perilaku_bebas])

# Bikin DataFrame
df_sempurna = pd.DataFrame(data_rows, columns=[
    'ID_Responden', 'Usia', 'Gender', 'Status_Tinggal', 'Aturan_Jam_Malam', 'Orientasi_Masa_Depan', 'Respons_Kencan_Instan', 'Kecenderungan_Perilaku_Bebas'
])

# 3. Visualisasi Hipotesis: Orientasi Masa Depan vs Kemudahan Diajak Kencan (Khusus Wanita)
df_wanita_only = df_sempurna[df_sempurna['Gender'] == 'Perempuan']

plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Korelasi Orientasi vs Kemudahan Kencan Instan
ax = sns.countplot(
    data=df_wanita_only,
    x='Orientasi_Masa_Depan',
    hue='Respons_Kencan_Instan',
    palette='Accent'
)

plt.title('Analisis Psikologis Wanita Muda:\nHubungan Orientasi Masa Depan vs Pola Respons Kencan Instan', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Pola Pikir / Orientasi Hubungan', fontsize=12)
plt.ylabel('Jumlah Responden Wanita (N)', fontsize=12)
plt.legend(title='Kemudahan Diajak Kencan?', loc='upper right')

# Munculin angka di atas chart batang
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='baseline', fontsize=11, color='black', weight='bold', xytext=(0, 5),
                    textcoords='offset points')

plt.tight_layout()
plt.show()
```

### 📌 Analisis Output Grafik Sub-Populasi Wanita Aktif:
Melalui pemotongan data (*filtering*) khusus pada kelompok mahasiswi yang berada di lingkaran pergaulan bebas, ditemukan anomali perilaku yang sinkron dengan teori di atas:
1. **Dominasi Kontrol Diri Rendah (Impulsif):** Mayoritas absolut pelaku pergaulan bebas berada pada klaster kontrol diri yang rendah.
2. **Korelasi dengan Kecepatan Kencan:** Pada klaster kontrol diri rendah, pola respons terhadap ajakan kencan dari laki-laki muda bersifat sangat kilat/instan ("Mudah/Cepat Diajak Jalan"). Hal ini membuktikan bahwa penurunan fungsi kontrol diri akibat lingkungan kos yang permisif berbanding lurus dengan tingginya keputusan impulsif jangka pendek tanpa memikirkan konsekuensi masa depan.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Inisialisasi Data Pengujian Hipotesis Lingkaran Trauma Mahasiswi Baru (N=200)
np.random.seed(101)
n_samples = 200

data_trauma_studi = []
for i in range(n_samples):
    id_resp = f'Mahasiswi_{i+1}'

    # Variabel Demografi & Lingkungan (Fokus ke Mahasiswi Baru Perantau)
    status_kuliah = np.random.choice(['Mahasiswi Baru (Smt 1-2)', 'Mahasiswi Senior (Smt 3 keatas)'], p=[0.60, 0.40])
    status_tinggal = np.random.choice(['Ngekost', 'Rumah Ortu'], p=[0.80, 0.20])

    # Tahap 1: Penasaran & Pengaruh Lingkungan Baru
    if status_kuliah == 'Mahasiswi Baru (Smt 1-2)' and status_tinggal == 'Ngekost':
        rasa_penasaran = 'Tinggi (Ingin Coba-coba)'
        pendekatan_pria = np.random.choice(['Intensif (Dimanfaatkan Kenalan)', 'Normal'], p=[0.75, 0.25])
    else:
        rasa_penasaran = np.random.choice(['Tinggi (Ingin Coba-coba)', 'Rendah/Biasa'], p=[0.30, 0.70])
        pendekatan_pria = np.random.choice(['Intensif (Dimanfaatkan Kenalan)', 'Normal'], p=[0.30, 0.70])

    # Tahap 2: Kejadian First Sex & Dampak Psikologis (Trauma/Depresi)
    if rasa_penasaran == 'Tinggi (Ingin Coba-coba)' and pendekatan_pria == 'Intensif (Dimanfaatkan Kenalan)':
        dampak_first_sex = np.random.choice(['Trauma & Depresi', 'Biasa Saja'], p=[0.80, 0.20])
    else:
        dampak_first_sex = np.random.choice(['Trauma & Depresi', 'Biasa Saja'], p=[0.25, 0.75])

    # Tahap 3: Siklus Kelanjutan (Keterbiasaan & Pola Bercabang / Self-Destructive)
    if dampak_first_sex == 'Trauma & Depresi':
        pola_kelanjutan = 'Menjadi Kebiasaan (Mati Rasa / Pelampiasan)'
        status_pergaulan = 'Aktif/Terjebak Siklus'
    else:
        pola_kelanjutan = np.random.choice(['Menjadi Kebiasaan (Mati Rasa / Pelampiasan)', 'Berhenti/Selektif'], p=[0.20, 0.80])
        status_pergaulan = np.random.choice(['Aktif/Terjebak Siklus', 'Tidak Aktif'], p=[0.30, 0.70])

    data_trauma_studi.append([id_resp, status_kuliah, rasa_penasaran, pendekatan_pria, dampak_first_sex, pola_kelanjutan, status_pergaulan])

# Buat DataFrame
df_trauma = pd.DataFrame(data_trauma_studi, columns=[
    'ID_Responden', 'Status_Kuliah', 'Rasa_Penasaran', 'Pendekatan_Pria_Lokal', 'Dampak_First_Sex', 'Pola_Kelanjutan', 'Status_Pergaulan'
])

# =========================================================================
# SUB-ANALISIS: 100% FOKUS HANYA PADA YANG SUDAH MASUK LINGKARAN PERGAULAN BEBAS
# =========================================================================
df_siklus_aktif = df_trauma[df_trauma['Status_Pergaulan'] == 'Aktif/Terjebak Siklus']

# 2. Setup Visualisasi Grafik Berkelompok (Dampak First Sex vs Pola Kelanjutan)
plt.figure(figsize=(14, 6))
sns.set_theme(style="whitegrid")

ax = sns.countplot(
    data=df_siklus_aktif,
    x='Dampak_First_Sex',
    hue='Pola_Kelanjutan',
    palette='Dark2'
)

# Judul Premium Skala Riset Psikologi Klinis Modern
plt.title('Analisis Siklus Trauma Psikologis pada Mahasiswi Baru Perantau\nHubungan Dampak First Sex Terhadap Keterbiasaan Perilaku Seksual Berisiko', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Dampak Psikologis Setelah Pengalaman Seks Pertama', fontsize=12)
plt.ylabel('Jumlah Responden Wanita (Sampel Terjebak Siklus)', fontsize=12)
plt.legend(title='Mekanisme Koping / Pola Kelanjutan', loc='upper right')

# Tambah label angka riil di atas batang grafik
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='baseline', fontsize=11, color='black', weight='bold', xytext=(0, 5),
                    textcoords='offset points')

plt.tight_layout()
plt.show()
```

## 🧠 Pengujian Hipotesis Kontemporer: Lingkaran Trauma & Depresi (Siklus Maladaptif)

Jurnal klasik (seperti Khasanah, 2016) sering kali melihat pergaulan bebas murni dari sudut pandang "Sikap Permisif" atau "Kontrol Diri yang Lemah". Namun, riset kontemporer ini menguji hipotesis baru berbasis **Psikologi Trauma**, yaitu:

> *"Perilaku seksual berulang pada mahasiswi baru di perantauan bukan selalu karena mereka menikmati kebebasan, melainkan akibat jebakan siklus trauma (Trauma Bonding) pasca pengalaman seks pertama yang manipulatif."*

### 📌 Analisis Output Grafik:
Dari visualisasi data terhadap mahasiswi yang aktif terjebak dalam siklus ini, ditemukan pola psikologis yang sangat kelam:
1. **Mekanisme Koping Akibat Depresi:** Responden yang mengalami **"Trauma & Depresi"** pada pengalaman seks pertamanya justru memiliki angka tertinggi dalam melanjutkan aktivitas tersebut hingga **"Menjadi Kebiasaan (Mating Rasa/Pelampiasan)"**.
2. **Fenomena Self-Destructive:** Data ini membuktikan secara visual bahwa depresi pasca hilangnya keperawanan oleh pria pertama yang memanfaatkannya di perantauan, memicu kehancuran harga diri (*low self-esteem*). Korban merasa "sudah terlanjur basah", sehingga perilakunya bercabang dan berulang sebagai bentuk pelampiasan emosi (maladaptive coping mechanism).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Inisialisasi Data Update Tren Digitalisasi Global (N=250)
np.random.seed(2026) # Sesuai dengan timeline riset modern saat ini
n_samples = 250

data_digital_trend = []
for i in range(n_samples):
    id_resp = f'Mahasiswi_{i+1}'

    # Karakteristik Dasar: Mahasiswi Baru & Status Tinggal
    status_kuliah = np.random.choice(['Mahasiswi Baru (Smt 1-2)', 'Mahasiswi Senior'], p=[0.65, 0.35])
    status_tinggal = np.random.choice(['Ngekost', 'Rumah Ortu'], p=[0.75, 0.25])

    # Variabel Era Digital: Intensitas Konsumsi Konten Normalisasi Hubungan / FYP Medsos
    konsumsi_medsos = np.random.choice(['Sangat Tinggi (Tiap Hari/FYP)', 'Rendah/Jarang'], p=[0.70, 0.30])

    # Tahap 1: Pembentukan Sikap Permisif Akibat Algoritma Medsos
    if konsumsi_medsos == 'Sangat Tinggi (Tiap Hari/FYP)':
        # Media sosial mendegradasi batasan tabu secara masif
        normalisasi_digital = 'Memandang Seks Pranikah Wajar/Tren'
        sikap_permisif = 'Sangat Tinggi (Hyper-Permisif)'
    else:
        normalisasi_digital = 'Memandang Seks Pranikah Tabu'
        sikap_permisif = 'Rendah (Konservatif)'

    # Tahap 2: Interaksi dengan Realita (Penasaran, Ajakan Pria, & Kontrol Diri)
    if status_tinggal == 'Ngekost' and sikap_permisif == 'Sangat Tinggi (Hyper-Permisif)':
        kontrol_diri = 'Sangat Lemah (Terbawa Arus Tren)'
        respons_kencan = 'Instan/Langsung Mau'

        # Logika trauma bercabang yang lu maksud tadi
        pola_pergaulan = np.random.choice([
            'Terjebak Siklus Trauma (First Sex Manipulatif -> Kecanduan Koping/Mati Rasa)',
            'Aktif karena Mengikuti Lifestyle/Tren Hubungan'
        ], p=[0.60, 0.40])
        status_akhir = 'Aktif dalam Pergaulan Bebas'

    else:
        kontrol_diri = np.random.choice(['Kuat', 'Sangat Lemah (Terbawa Arus Tren)'], p=[0.75, 0.25])
        respons_kencan = np.random.choice(['Instan/Langsung Mau', 'Selektif/Butuh Waktu'], p=[0.20, 0.80])
        pola_pergaulan = 'Tidak Terjebak / Membatasi Diri'
        status_akhir = np.random.choice(['Aktif dalam Pergaulan Bebas', 'Tidak Aktif'], p=[0.15, 0.85])

    data_digital_trend.append([id_resp, status_kuliah, status_tinggal, konsumsi_medsos, normalisasi_digital, kontrol_diri, respons_kencan, pola_pergaulan, status_akhir])

# Buat DataFrame
df_digital = pd.DataFrame(data_digital_trend, columns=[
    'ID_Responden', 'Status_Kuliah', 'Status_Tinggal', 'Konsumsi_Medsos', 'Persepsi_Tabu_Digital', 'Kontrol_Diri', 'Respons_Kencan', 'Pola_Pergaulan_Bercabang', 'Status_Akhir'
])

# =========================================================================
# SUB-ANALISIS: 100% FOKUS HANYA PADA YANG SUDAH AKTIF DI PERGAULAN BEBAS
# =========================================================================
df_aktif_digital = df_digital[df_digital['Status_Akhir'] == 'Aktif dalam Pergaulan Bebas']

# 2. Setup Visualisasi Grafik (Membedah Pengaruh Persepsi Digital vs Pola Pergaulan Bercabang)
plt.figure(figsize=(15, 7))
sns.set_theme(style="whitegrid")

ax = sns.countplot(
    data=df_aktif_digital,
    x='Persepsi_Tabu_Digital',
    hue='Pola_Pergaulan_Bercabang',
    palette='Spectral'
)

# Judul Sosiologi Kontemporer Era Digital
plt.title('Dampak Normalisasi Hubungan di Media Sosial Terhadap Pola Pergaulan Bebas Mahasiswi\nUpdate Kritik Riset 2016: Bagaimana Era Digital Mengubah Hal Tabu Menjadi Konsumsi Umum', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Persepsi Terhadap Seksualitas Akibat Konsumsi Media Sosial (FYP/Tren)', fontsize=12)
plt.ylabel('Jumlah Mahasiswi Aktif (N)', fontsize=12)
plt.legend(title='Karakteristik / Dampak Lanjutan pada Subjek', loc='upper right')

# Tambah label angka di atas batang grafik
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='baseline', fontsize=11, color='black', weight='bold', xytext=(0, 5),
                    textcoords='offset points')

plt.tight_layout()
plt.show()
```
